```python
#==============================================
ni = 0
n  = 9
xx1 = np.linspace( np.min(x_[ni:n]), np.max(x_[ni:n]), len(x_[ni:n]))
p_lagrange1 = lagrange(x_[ni:n], y_[ni:n])
#==============================================
ni = 8
n  = 22
xx2 = np.linspace( np.min(x_[ni:n]), np.max(x_[ni:n]), len(x_[ni:n]))
p_lagrange2 = lagrange(x_[ni:n], y_[ni:n])
#==============================================

xx = np.linspace( np.min(x_), np.max(x_), 100)
S_cubico =  interpolate.interp1d(x_, y_, kind = "cubic")
#S_cubico = interpolate.CubicSpline(x_, y_ ) #Las parejas deben de estar en orden para emplear esta rutina
plt.figure(figsize=(10, 8))
plt.imshow(imagen)
plt.plot(x_, y_,  'ro', markersize = 10)
plt.plot(xx1, p_lagrange1(xx1), "y-", lw=3, label='Interp. Lagrange')
plt.plot(xx2, p_lagrange2(xx2), "y-", lw=3, label='Interp. Lagrange')
#plt.plot(xx, S_cubico(xx),   "b-", lw=3, label = u'Splines cúbicos')
plt.ylim(300, 0)
plt.xlim(300, 800)
plt.legend(loc='lower left')
plt.show()
```
