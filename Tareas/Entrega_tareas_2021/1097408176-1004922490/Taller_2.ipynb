## Campo eléctrico 
1. Elaborar un programa para calcular el campo eléctrico producido por un dipolo eléctrico, realizar una representación del campo para todo el dominio 
elegido para calcular el campo. Para encontrar el campo, se sugiere hacer lo siguiente:

- a. Encontrar el campo en un punto producido por dos cargas q1 y q2
- b. Emplee el comando  plt.quiver(x, y, Ex, Ey) para mostrar el campo producido por q1 y q2. Puedes intentarlo con el comando, plt.stremplot.
- c. Generalice el problema para varios puntos en el espacio.

Se recomienda emplear numpy array. 

```python
#importando librerias 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#pedir al usuario ingresar la posicion y el valor de la carga
q_a=int(input("ingrese el valor de la carga 1 "))
q_b=int(input("ingrese el valor de la carga 2 "))
q_ax=int(input("ingrese el valor de la posicion en x de la carga 1 "))
q_ay=int(input("ingrese el valor de la posicion en y de la carga 1 "))
q_bx=int(input("ingrese el valor de la posicion en x de la carga 2 "))
q_by=int(input("ingrese el valor de la posicion en y de la carga 2 "))


k=1
def campoelec(q, r): #función para hallar magbitud del campoelectrico de cada carga
  return (k*q)/r**2
#función para hallar el campo de las cargas como vector
def ccarga (posicionq, posicionp, q): 
  x =  posicionp - posicionq 
  magx = np.sqrt(x[0]**2 + x[1]**2)
  if magx!=0:
    Ur = x/magx #vector unitario en direccion del campo
    E = campoelec(q, magx ) * Ur #valor del campo
    E_y = E[1]
    E_x = E[0]
    return E_x, E_y #vectores del campo 
#espacio para hallar el campo 
grilla1 = np.arange(-2, 3, 0.5) 
longitud1 = len(grilla1)
grilla2 = np.arange(-2, 3, 0.5)
longitud2 = len(grilla2)
posicionp = np.zeros((longitud1*longitud2 , 2))
j = 0
for x_ in (grilla1):
  for y_ in (grilla2):
    posicionp[j] = np.array([x_, y_])
    j=j+1 

r_source = [np.array([q_ax, q_ay]), np.array([q_bx, q_by])] #posición de las cargas
q= [q_a, q_b]# introducir valores de las cargas del dipolo
E = np.zeros((2,longitud1*longitud2, 2))

for h in range(len(q)):
  for i in range (0, len(posicionp)):
    E[h][i] = ccarga(r_source[h], posicionp[i], q[h])

Etotal = sum(E)
#graficar 
plt.plot(r_source[0][0], r_source[0][1], "o")
plt.plot(r_source[1][0], r_source[1][1], "o")
plt.title("Líneas de campo")
plt.xlabel('$X(m)$')
plt.ylabel('$Y(m)$')
plt.quiver(posicionp[:, 0], posicionp[:, 1], Etotal[:, 0], Etotal[:, 1])
```

## Movimiento parabólico.
2. Un jugador de baloncesto desea saber si encestará su tiro dependiendo de la velocidad inicial, ángulo de lanzamiento y posición respecto a la canasta. 

Elabore un programa en el que dada una distancia, velocidad inicial y  ángulo, determine si hay cesta. El modelo será bajo partícula, sin la rotación del balón. 

Se recomienda emplear numpy array.

```python
h=2.75 #altura de la cesta
da=0.20 #diametro del arco
xi = float(input("ingrese la distancia inicial en el eje x en m "))
yi = float(input("ingrese la distancia inicial en el eje y en m "))
gradoi = float(input("ingrese el grado "))
vi= float(input("ingrese el valor de la velocidad inicial en m/s "))
#regla de 3 para pasar de grados a radianes
g=9.8
theta= (gradoi*np.pi)/180
H=h-yi #altura
#ecuaciones cinematica
def Y(x): y=(np.tan(theta)*x)-((g*(x**2))/(2*(vi**2)*(np.cos(theta)**2)))+yi; return y
X=((vi**2)*np.sin(2*theta))/g #ecuacion alcance horizontal
altura = False
distancia = False
#saber si hay en cesta 
if abs(Y(xi) - H) < 1.0e-5:
  altura = True
if da>X>0:
  distancia = True
if altura and distacia == True:
  print("Hay cesta")
else:
  print("No hay cesta")

#Se hace la grafica

plt.plot(xi,h,"r*",label="Canasta")
plt.scatter(0,yi,label="Lanzamiento",color='pink')
plt.axhline(0,0,xi)
plt.axvline(xi,0,yi)
plt.title("Tiro parabòlico")
X = np.linspace(0,6,50)
plt.plot(X,Y(X),'--',label='Trayectoria', color="purple")
plt.xlabel('$X(m)$')
plt.ylabel('$Y(m)$')
plt.legend();



```
