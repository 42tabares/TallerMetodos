```python
#Densidad
rho_0 = 1.29 #kg/m^3
#Parámetro
lambd = 7482.2 #m
```

```python
#NUMERAL 1 ,  DENSIDAD EN FUNCIOND DE LA ALTURA
densidad = lambda y: rho_0 * np.exp(-(y/lambd))

#Gráfico
x = np.linspace(0,30000,100)
plt.figure()
plt.plot(x , densidad(x),"g-")
plt.xlabel("Altura [m]")
plt.ylabel("Densidad [kg/m^3]")
plt.title("Densidad exponencial")
plt.grid()
```

```python
#NUMERAL 2, parámetros

m = 60 #kg
A = 0.3 #m^2
h = np.linspace(30,70,5) #km
g = 9.8 #m/s^2
k = 0.5
k_0 = 0.5 * rho_0 * A * k
h = np.linspace(30,70,5)
#Parametros
params = [m, k_0, lambd, g]
```

```python
# TRANSFORMAMOS AL EDO EN 2 EDO
def func(x,t,params):
  g = params[3]
  m = params[0]
  k_0 = params[1]
  lamd = params[2]

  dxdt = x[1]
  dvdt = -g + k_0 * ((x[1]**2)/m) * np.exp (-x[0]/lambd)
  return [dxdt, dvdt]

t = np.linspace(0, 200, 1000) #Tiempo de integración


  
```

```python

#Condiciones Iniciales
v_0 = 0
x_0 = np.array([30000,40000,50000,60000,70000])
#Para cada uno de los valores de altura
x0 = np.array([x_0[0],v_0])
x_a = integrate.odeint(func, x0 ,t, args = (params, ))

x0 = np.array([x_0[1],v_0])
x_b = integrate.odeint(func, x0 ,t, args = (params, ))

x0 = np.array([x_0[2],v_0])
x_c = integrate.odeint(func, x0 ,t, args = (params, ))

x0 = np.array([x_0[3],v_0])
x_d = integrate.odeint(func, x0 ,t, args = (params, ))

x0 = np.array([x_0[4],v_0])
x_e = integrate.odeint(func, x0 ,t, args = (params, ))

```

```python
#NUMERAL 3, GRAFICO
x1_a = x_a[:,0]
x2_a = x_a[:,1]

x1_b = x_b[:,0]
x2_b = x_b[:,1]

x1_c = x_c[:,0]
x2_c = x_c[:,1]

x1_d = x_d[:,0]
x2_d = x_d[:,1]

x1_e = x_e[:,0]
x2_e = x_e[:,1]


plt.figure()
plt.subplot(121)
plt.plot(t,x1_a, label = "30km") #posicion en funcion del tiempo
plt.plot(t,x1_b, label = "40km")
plt.plot(t,x1_c, label = "50km")
plt.plot(t,x1_d, label = "60km")
plt.plot(t,x1_e, label = "70km")
plt.xlabel("t [s]")
plt.ylabel("x(t) [m]")
plt.title("Posición en función del tiempo")
plt.grid()
plt.legend()
plt.subplot(122)
plt.plot(t,x1_a, label = "30km") #velociad en funcion del tiempo
plt.plot(t,x1_b, label = "40km")
plt.plot(t,x1_c, label = "50km")
plt.plot(t,x1_d, label = "60km")
plt.plot(t,x1_e, label = "70km")
plt.xlabel("t [s]")
plt.ylabel("v(t)[m/s]")
plt.title("Velocidad en función del tiempo")
plt.grid()
plt.legend()
```
