```python
#Uso de odeint 
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pylab as plt
import numpy as np
```

\begin{equation}
m_1 \frac{ \mathrm{d}^2 x_1 }{\mathrm{d}t^2} = -k_1 x_1 + k_2 (x_2-x_1)
\end{equation}


\begin{equation}
m_2 \frac{ \mathrm{d}^2 x_2 }{\mathrm{d}t^2} = -k_3 x_2 - k_2 (x_2-x_1)
\end{equation}


$$\frac{ \mathrm{d}^2 x_1 }{\mathrm{d}t^2} = -\frac{k_1 }{m_1}x_1 + \frac{k_2 }{m_1} (x_2-x_1)$$
$$\frac{ \mathrm{d}^2 x_1 }{\mathrm{d}t^2} = -\alpha_1x_1 + \beta_1 (x_2-x_1)$$
\
$$\frac{ \mathrm{d}^2 x_2 }{\mathrm{d}t^2} = -\frac{k_3 }{m_2}x_2 - \frac{k_2 }{m_2}(x_2-x_1)$$
$$\frac{ \mathrm{d}^2 x_2 }{\mathrm{d}t^2} = -\alpha_2x_2 - \beta_2 (x_2-x_1)$$

```python
def func(x,t,params):
  a1=params[0]
  a2=params[1]
  b1=params[2]
  b2=params[3]

  dx1=x[2]
  dv1=-a1*x[0]+b1*(x[1]-x[0])
  dx2=x[3]
  dv2=-a2*x[1]-b2*(x[1]-x[0])

  return (dx1,dx2,dv1,dv2)
```

```python
x0=np.array([-2,1,-3,10])#posiciones y velocidades iniciales
t =np.linspace(0, 500, 10000) 

params=(0.1,0.2,0.3,0.2)
x= odeint(func, x0, t, args=(params, ) )


x1 = x[:,0] 
v1 = x[:,1] 
x2 = x[:,2]
v2 = x[:,3]

fig1=plt.figure(figsize=(20,10))
plt.subplot(221)
plt.plot(t, x1)
plt.xlabel("t")    
plt.ylabel("x1(t)")
plt.grid()
plt.title('x1')

plt.subplot(222)
plt.plot(t,v1) 
plt.xlabel("t")   
plt.ylabel("x2(t)") 
plt.grid()
plt.title('x2')

plt.subplot(223)
plt.plot(t,x2) 
plt.xlabel("t")   
plt.ylabel("v1(t)")
plt.grid()
plt.title('v1')

plt.subplot(224)
plt.plot(t,v2)  
plt.xlabel("t")
plt.ylabel('v2(t)')
plt.grid()
plt.title('v2')

plt.show()
```
