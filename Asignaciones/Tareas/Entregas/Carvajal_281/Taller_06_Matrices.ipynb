#AUTOVALORES Y AUTOVECTORES

Se sabe que un sistema físico es gobernado por las ecuaciones,



$$
\begin{matrix}
x' &=& 2x&-y&+z, \\
y' &=&   &3y&-1z, \\
z' &=& 2x&+y&+3z,
\end{matrix}
$$

donde $x',y',z'$ son las derivadas, en forma vectorial esto se puede representar como $\mathbf{r}'=A\mathbf{r}$ que tiene solución analítica general, $\mathbf{r} = e^{At} \mathbf{r}(0)$.<br> 
a) Diagonalize $A$.<br> 

b) Use el hecho de que $f(A) =U f(D) U^{-1}$ para encontrar $x(t),y(t),z(t)$ en el intervalo de tiempo $[0,2]$.<br>.
c) Grafique $x(t),y(t),z(t)$ (use la condición inicial $\mathbf{r}(0)=(1,1,2))$.<br> 
d) Compare el resultado a la solución analítica,

$$
\begin{align}
2x & = x(0)(e^{2t}(1+e^{2t}-2t)) + y(0) (-2te^{2t}) + z(0)(e^{2t}(-1+e^{2t})), \\
2y & = x(0)(-e^{2t}(-1+e^{2t}-2t)) + y(0)(2(t+1)e^{2t}) + z(0)(-e^{2t}(-1+e^{2t})), \\
2z & = x(0)(e^{2t}(-1+e^{2t}+2t)) + y(0)(2te^{2t}) + z(0)(e^{2t}(1+e^{2t})).
\end{align}
$$







#PROCEDIMIENTO#


1.   Para diagonalizar la matriz A se hace uso de la ecuación $A= UDU^{-1}$ donde P son los autovectores de la matriz , por lo tanto  $D= U^{-1}AU$ 
2.   Para encontrar $x(t),y(t),z(t)$ en el intervalo de tiempo $[0,2]$.<br>






```python
import numpy as np
import sympy as sp
A= np.array([[2,-1,1],[0,3,-1],[2,1,3]])
Diag=np.linalg.eig(A) 
val= Diag[0]
Pvects= Diag[1]
Pvecinv= np.linalg.inv(Pvects)
print(f"El determinante de la matriz es: {(np.linalg.det(Pvects))}, por lo tanto la función es invertible",)

D= Pvecinv@A@Pvects
print("La matriz diagonalizada D es:")
sp.Matrix(D)
```

```python
a=[]
t=np.linspace(0,2,10)
for t_ in t:
  di = np.exp(val*t_)
  fd = Pvects@np.diag(di)@Pvecinv
  a.append(fd)
a= np.array(a)
a=a.real
```

```python
r0= (1,1,2)
ci= (a[0]@r0).real

print(f"Las condiciones iniciales son:  \nX(0)={ci[0]}\nY(0)= {ci[1]} \nZ(0)={ci[2]}\n")

ct= []
for i in range(len(t)):
  ci= a[i]@r0
  ct.append(ci)

ct= np.array(ct)
ct= ct.real

print("La matriz de los X(t),Y(t),Z(t) es:")
sp.Matrix(ct)
```
