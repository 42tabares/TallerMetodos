Escribe un programa que imprima los valores de e^(-x)


```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
f= lambda x: np.exp(-x)
```

```python
x= np.linspace(-5,5,100)
y=f(x)
plt.plot(x,y)
```

\begin{equation}
e^{-x}=\sum_{n=0}^{N}  (-1)^n \frac{x^{n}}{n!}
\end{equation}

```python
def exponencial1(x, n): 
    _exp = 0
    for i in range(0, n):
      num  = ((-1) ** i ) * x ** i
      den  = mt.factorial(i)
      _exp = _exp + num / den 
    return _exp

def exponencial2(x, n):
  _exp = 0
  for i in range(0, n):
    num  =  x ** i
    den  = mt.factorial(i)
    _exp = _exp + num / den 
  return 1/_exp

```

Comparación entre los valores de la exponencial en 10 

```python
N = 100
x1 = 10
e1 = exponencial1(x1,N )
e2 = exponencial2(x1,N )
e_exa = np.exp(-x1)

print(f'{e1}')
print(f'{e2}')
print(f'{e_exa}')

```

Comparación entre las graficas teoricas y computacionales

```python
x = np.linspace(-5,1,100)
n=100
plt.figure()
plt.plot(x,exponencial1(x,n),"o",label='Computacional 1')
plt.plot(x,exponencial2(x,n),"-",label='Computacional 2')
plt.plot(x,y,label='Teorico')
plt.legend()

```

```python
x = 10
N = np.arange(3, 20, 1)
exp1_ = np.zeros(len(N))
exp2_ = np.zeros(len(N))

for i,n in  enumerate(N):
  exp1_[i] =  exponencial1(x = x, n = n)
  exp2_[i] =  exponencial2(x = x, n = n)
plt.figure()
plt.plot(N, exp1_,"r-", label="Serie alternante exp(x)"   )
plt.plot(N, exp2_,"b-",label="1/exp(x)"   )
plt.legend()
plt.show()
```

Encontrar tres formas diferentes de expresar y calcular la evolución de la misma condición inicial después de cientos de iteraciones. Para este problema, será extremadamente útil ver sus resultados gráficamente; construya listas de números y llame la libreria matplotlib.

```python
r = 2.9
xo = 0.6
N = 25
X_= np.zeros(N)
t = np.arange(0, N, 1)
for i in range(0, N):
  x = r*xo*(1-xo)
  xo = x
  X_[i] = xo
  #print(f"{xo}")
plt.figure()
plt.plot(t,X_,label=f"r={r}")
plt.xlabel("t(ua)")
plt.ylabel("x")
plt.legend()
```

```python
r = 1.9
xo = 0.6
N = 25

def poblacion(r,xo,N):

  X_= np.zeros(N)
  t = np.arange(0, N, 1)
  for i in range(0, N):
    x = r*xo*(1-xo)
    xo = x
    X_[i] = xo
    #print(f"{xo}")
  
  dicc = {"x":X_, "t":t}
  return dicc

r = 2.9
xo = 0.6
N = 25
d1 = poblacion(r,xo,N)

r=3.5
xo = 0.5
N = 25
d2 = poblacion(r,xo,N)
```

```python
plt.plot(d1["t"],d1["x"],label='2.9' )
plt.plot(d2["t"],d2["x"],label='3.5')
plt.legend()
```
